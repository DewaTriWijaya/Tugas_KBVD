## Deskripsi Parameter

Berikut adalah penjelasan mengenai berbagai parameter yang digunakan dalam pemantauan kualitas udara dan kondisi lingkungan:

### 1. **PM1 (Particulate Matter 1)**
   - **Deskripsi**: PM1 adalah partikel halus di udara yang lebih kecil dari 1 mikron (µm). Partikel ini sangat kecil dan dapat masuk ke dalam saluran pernapasan dalam.
   - **Sumber**: Biasanya berasal dari pembakaran bahan bakar fosil, asap kendaraan, dan proses industri.
   - **Dampak Kesehatan**: Dapat menyebabkan gangguan pernapasan, penyakit paru-paru, dan meningkatkan risiko penyakit jantung.

### 2. **PM10 (Particulate Matter 10)**
   - **Deskripsi**: PM10 adalah partikel debu di udara dengan ukuran lebih kecil dari 10 mikron. Meskipun lebih besar dari PM1, PM10 masih bisa masuk ke saluran pernapasan dan menyebabkan masalah kesehatan.
   - **Sumber**: Aktivitas manusia seperti konstruksi, pembakaran biomassa, dan polusi industri, serta fenomena alam seperti debu dan pasir.
   - **Dampak Kesehatan**: Paparan jangka panjang dapat menyebabkan gangguan pernapasan, asma, bronkitis, dan penyakit paru-paru lainnya.

### 3. **PM2.5 (Particulate Matter 2.5)**
   - **Deskripsi**: PM2.5 adalah partikel halus dengan ukuran lebih kecil dari 2.5 mikron. Karena ukurannya yang sangat kecil, partikel ini dapat masuk jauh ke dalam saluran pernapasan, bahkan ke dalam aliran darah.
   - **Sumber**: Dihasilkan dari pembakaran bahan bakar, asap kendaraan, aktivitas industri, serta kebakaran hutan.
   - **Dampak Kesehatan**: Memicu gangguan pernapasan, penyakit jantung, kanker paru-paru, dan gangguan lainnya pada sistem pernapasan.

### 4. **Relative Humidity (Kelembaban Relatif)**
   - **Deskripsi**: Kelembaban relatif mengukur jumlah uap air yang ada di udara dibandingkan dengan jumlah maksimum yang dapat dipegang udara pada suhu yang sama.
   - **Sumber**: Dipengaruhi oleh suhu udara, cuaca, serta faktor lingkungan.
   - **Dampak Kesehatan**: Kelembaban tinggi dapat meningkatkan pertumbuhan jamur dan bakteri, sedangkan kelembaban rendah dapat menyebabkan iritasi pada saluran pernapasan.

### 5. **Temperature (Suhu)**
   - **Deskripsi**: Suhu udara mengukur tingkat panas atau dinginnya udara, yang sangat berpengaruh terhadap kondisi cuaca dan kenyamanan manusia.
   - **Sumber**: Dipengaruhi oleh faktor alami seperti waktu siang dan malam, musim, serta posisi geografis.
   - **Dampak Kesehatan**: Suhu yang sangat panas atau dingin dapat menyebabkan gangguan kesehatan seperti heatstroke, hipotermia, atau dehidrasi.

### 6. **UM003**
   - **Deskripsi**: UM003 kemungkinan merujuk pada sensor atau perangkat tertentu untuk mengukur kualitas udara atau parameter lingkungan lainnya. Perangkat ini dapat digunakan untuk memantau salah satu atau lebih dari parameter-parameter kualitas udara yang telah disebutkan.
   - **Sumber**: Data ini mungkin berasal dari stasiun pemantauan kualitas udara atau perangkat pengukur tertentu yang terhubung ke sistem pemantauan lingkungan.
   - **Dampak Kesehatan**: Dampaknya bergantung pada jenis parameter yang diukur oleh perangkat UM003.

### Kesimpulan:
- **PM1**, **PM10**, dan **PM2.5** adalah jenis partikel halus di udara yang dapat menyebabkan masalah kesehatan serius, terutama pada sistem pernapasan.
- **Kelembaban relatif** dan **suhu** merupakan faktor lingkungan yang mempengaruhi kenyamanan dan kesehatan.
- **UM003** merujuk pada perangkat pemantauan yang mengukur parameter lingkungan, tetapi informasi lebih lanjut dibutuhkan untuk menjelaskan fungsi spesifiknya.

```python
# Convert 'datetimeLocal' to datetime format and extract the hour
df['datetimeLocal'] = pd.to_datetime(df['datetimeLocal'])
df['hour'] = df['datetimeLocal'].dt.hour

# Group by hour and parameter, and calculate the mean value
hourly_avg = df.groupby(['hour', 'parameter'])['value'].mean().reset_index()

hourly_avg.tail()
```

```python
# Filter data untuk parameter PM1 dan rentang jam tertentu (misal, jam 0-23)
start_hour, end_hour = 6, 18
filtered_data = hourly_avg[(hourly_avg['parameter'] == 'pm1') & 
                           (hourly_avg['hour'] >= start_hour) & 
                           (hourly_avg['hour'] <= end_hour)]

# Create the plot
plt.figure(figsize=(10, 6))

# Line chart with color for similarity and smoothness (Gestalt: Similarity, Continuation)
sns.lineplot(data=filtered_data, x='hour', y='value', marker='o', color='blue', linewidth=2, label="PM1 Levels")

# Highlight maximum value using color and size (Preattentive: Color and Size)
max_value = filtered_data['value'].max()
max_hour = filtered_data[filtered_data['value'] == max_value]['hour'].iloc[0]
plt.scatter(x=max_hour, y=max_value, color='red', s=120, label=f'Max Value ({max_value} µg/m³)')

# Add shaded areas for Enclosure principle (Gestalt: Enclosure)
plt.axvspan(start_hour, 12, color='lightblue', alpha=0.2, label='Morning Hours')
plt.axvspan(12, end_hour, color='lightyellow', alpha=0.2, label='Afternoon Hours')

# Customizing x-axis and y-axis for Proximity (Gestalt: Proximity)
plt.xticks(range(start_hour, end_hour + 1), fontsize=10)
plt.yticks(fontsize=10)

# Add title, labels, and grid (Gestalt: Continuation)
plt.title('Hourly Average PM1 Concentration', fontsize=14, fontweight='bold')
plt.xlabel('Hour', fontsize=12)
plt.ylabel('PM1 Concentration (µg/m³)', fontsize=12)
plt.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)

# Add a horizontal line to indicate a threshold for visual aid (optional enhancement)
threshold = 66
plt.axhline(y=threshold, color='black', linestyle='--', linewidth=1, label='Threshold (66 µg/m³)')

# Add legend for clarity
plt.legend(fontsize=10, loc='upper right')

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```

```python
df = pd.read_csv('data/openaq_depok_clean.csv')

# Convert the 'datetimeLocal' column to datetime format
df['datetimeLocal'] = pd.to_datetime(df['datetimeLocal'])

# Extract the month and year for grouping purposes
df['month_year'] = df['datetimeLocal'].dt.to_period('M')

# Identify the maximum and minimum values per month for each parameter
extreme_values = df.groupby(['month_year', 'parameter'])['value'].agg(['max', 'min']).reset_index()

# Display the results
extreme_values.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up visual style to align with principles of Gestalt and preattentive attributes
sns.set_theme(style="whitegrid")

# Prepare data for visualization (example for one parameter - 'pm10')
pm10_data = extreme_values[extreme_values['parameter'] == 'pm10']

# Create a bar chart with annotations for max and min values
fig, ax = plt.subplots(figsize=(12, 6))

# Plot max and min values
bar_width = 0.4
x = range(len(pm10_data))
ax.bar(x, pm10_data['max'], bar_width, label='Max', color='#4CAF50')
ax.bar([i + bar_width for i in x], pm10_data['min'], bar_width, label='Min', color='#FFC107')

# Add annotations for clarity
for i, val in enumerate(pm10_data['max']):
    ax.text(i, val + 5, f"{val:.1f}", ha='center', va='bottom', fontsize=10, color='#2E7D32')
for i, val in enumerate(pm10_data['min']):
    ax.text(i + bar_width, val - 10, f"{val:.1f}", ha='center', va='top', fontsize=10, color='#FF6F00')

# Add labels and titles
ax.set_title('Extreme Values for PM10 Over Time', fontsize=16, weight='bold', pad=20)
ax.set_xlabel('Month-Year', fontsize=12)
ax.set_ylabel('Value (µg/m³)', fontsize=12)
ax.set_xticks([i + bar_width / 2 for i in x])
ax.set_xticklabels(pm10_data['month_year'].astype(str), rotation=45, ha='right', fontsize=10)
ax.legend(title='Value Type', fontsize=10, loc='upper right')

# Gestalt Principles: Enclosure (clear gridlines) and Proximity (grouped bars)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('data/openaq_depok_clean.csv')

# Convert the 'datetimeLocal' column to datetime
data['datetimeLocal'] = pd.to_datetime(data['datetimeLocal'])

# Extract year
data['year'] = data['datetimeLocal'].dt.year

# Group by parameter and year, then calculate min and max
summary = data.groupby(['parameter', 'year'])['value'].agg(['min', 'max']).reset_index()

# Focus on a single parameter (e.g., 'pm25')
parameter_to_plot = 'pm25'
summary = summary[summary['parameter'] == parameter_to_plot]

# Plotting
plt.figure(figsize=(14, 8))
sns.set_theme(style="whitegrid")

# Plot min and max values for the selected parameter
plt.plot(summary['year'], summary['max'], label=f'{parameter_to_plot} (Max)', marker='o', linestyle='-', color='blue')
plt.plot(summary['year'], summary['min'], label=f'{parameter_to_plot} (Min)', marker='o', linestyle='--', color='red')

# Customize the plot
plt.title(f'Yearly Min and Max Values for {parameter_to_plot.upper()} in Depok', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Value (Units)', fontsize=14)  # Replace "Units" with the appropriate unit
plt.xticks(rotation=45)
plt.legend(title='Value Type', loc='upper left', fontsize=10)
plt.tight_layout()

# Highlight the highest and lowest values with annotations
max_row = summary.loc[summary['max'].idxmax()]
min_row = summary.loc[summary['min'].idxmin()]
plt.text(max_row['year'], max_row['max'], f'{max_row["max"]:.2f}', color='blue', fontsize=9)
plt.text(min_row['year'], min_row['min'], f'{min_row["min"]:.2f}', color='red', fontsize=9)

# Show the plot
plt.show()
```

5. Informasi Potensi Dampak Kesehatan dengan mencari nilai tertinggi dari setiap parameter, dan parameter apa yang tinggi?
    - Metode         : Analisis perbandingan nilai rata-rata tertinggi parameter terhadap standar atau ambang batas kualitas udara.
    - Grafik         : Tree Map atau Bubble Chart untuk menunjukkan kontribusi setiap parameter terhadap kualitas udara.
    - Jenis Analisis : Explanatory.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'data/openaq_depok_clean.csv'  # Ganti dengan path file Anda
data = pd.read_csv(file_path)

# Pastikan dataset memiliki kolom 'parameter' dan 'value'

# Ambang batas kualitas udara (contoh, sesuaikan dengan standar lokal/internasional)
thresholds = {
    'pm25': 25,   # Ambang batas PM2.5 (μg/m3, WHO)
    'pm10': 50,   # Ambang batas PM10 (μg/m3, WHO)
    'no2': 40,    # Ambang batas NO2 (μg/m3, WHO)
    'o3': 100,    # Ambang batas O3 (μg/m3, WHO)
    'co': 4       # Ambang batas CO (mg/m3, WHO)
}

# Hitung nilai rata-rata dan maksimum untuk setiap parameter
summary = data.groupby('parameter')['value'].agg(['mean', 'max']).reset_index()

# Tambahkan kolom ambang batas
summary['threshold'] = summary['parameter'].map(thresholds)

# Filter only valid parameters with thresholds
summary = summary[summary['threshold'].notnull()]

# Tambahkan kolom rasio terhadap ambang batas
summary['ratio_to_threshold'] = summary['max'] / summary['threshold']

# Bubble Chart
plt.figure(figsize=(12, 6))
sns.set_theme(style="whitegrid")

# Ukuran gelembung berdasarkan rasio terhadap ambang batas
bubble_sizes = summary['ratio_to_threshold'] * 1000  # Adjust size scale

# Scatter plot
plt.scatter(
    summary['parameter'], 
    summary['max'], 
    s=bubble_sizes, 
    alpha=0.7, 
    color='skyblue', 
    edgecolors='black'
)

# Tambahkan teks pada gelembung
for i in range(len(summary)):
    plt.text(
        summary['parameter'].iloc[i], 
        summary['max'].iloc[i], 
        f"{summary['parameter'].iloc[i]}\n({summary['max'].iloc[i]:.1f})", 
        ha='center', 
        fontsize=9
    )

# Customize plot
plt.title('Bubble Chart: Potensi Dampak Kesehatan Berdasarkan Parameter', fontsize=16)
plt.xlabel('Parameter', fontsize=14)
plt.ylabel('Nilai Maksimum', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(visible=True, linestyle='--', alpha=0.5)

# Show the plot
plt.tight_layout()
plt.show()
```
